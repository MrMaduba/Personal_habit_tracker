```python

# Function to validate username
def validate_username(username):
    if (len(username) >= 8 and
        re.search(r'\d', username) and
        re.search(r'[!@#$%^&*(),.?":{}|<>]', username)):
        return True
    return False

# Function to validate password
def validate_password(password):
    if (len(password) >= 8 and
        re.search(r'\d', password) and
        re.search(r'[!@#$%^&*(),.?":{}|<>]', password)):
        return True
    return False

# User input for username with validation
while True:
    username = input("Please enter a username (must be at least 8 characters, numbers and a special character inclusive): ")
    if not validate_username(username):
        print("Username must be at least 8 characters long, include at least one number and one special character.")
    else:
        print("Username is successfully set!")
        break

# User password requirement using while loop
while True:
    password = input("Please enter your password (must be at least 8 characters, numbers and a special character inclusive): ")
    password_confirm = input("Please re-enter your password: ")

    if password != password_confirm:
        print("Passwords do not match. Please try again.")
    elif not validate_password(password):
        print("Password must be at least 8 characters long, include at least one number and one special character.")
    else:
        print("Password successfully set!")
        break

# Defining the function to track habits
def breaking_habit(name_of_habit, start_date, daily_cost, minutes_wasted, wage_per_hour):
    # Personal info and inputs
    personal_goal = 66

    # Total time elapsed since start date and now in seconds
    time_elapsed = (datetime.now() - start_date).total_seconds()

    # Convert time stamp to hours and days
    hours = round(time_elapsed / 60 / 60, 2)
    days = round(hours / 24, 2)

    # Additional info
    money_saved = daily_cost * days
    minutes_saved = round(days * minutes_wasted)
    total_money_saved = f'${round(money_saved + minutes_saved / 60 * wage_per_hour, 2)}'
    days_left = round(personal_goal - days)

    # Change hours to days
    if hours > 72:
        hours_display = str(days) + ' days'
    else:
        hours_display = str(hours) + ' hours'

    # Feedback based on whether the habit is cleared
    if days >= personal_goal:
        feedback = f"Congratulations! You've broken the habit of '{name_of_habit}'."
    else:
        feedback = f"Keep going! You're on your way to breaking the habit of '{name_of_habit}'. Just {days_left} days to go!"

    return {
        'Habit': name_of_habit,
        'Time since tracking started': hours_display,
        'Remaining days to break habit': days_left,
        'Minutes saved so far': minutes_saved,
        'Money saved so far': total_money_saved,
        'Feedback on habit status': feedback
    }

# User input for wage per hour with validation using while loop and break
while True:
    try:
        wage_per_hour = float(input("Enter your wage per hour (in USD): "))
        if wage_per_hour < 0:
            print("Wage cannot be negative. Please enter a valid wage.")
            continue
        break  # Exit the loop if the input is valid
    except ValueError:
        print("Invalid input. Please enter a numeric value for wage per hour.")

# User input for habits using while loop, try and break
habits = []
while True:
    name_of_habit = input("Enter the name of the habit (or type 'exit' to finish): ")
    if name_of_habit.lower() == 'exit':
        break

    # Input for start date with validation using while loop, try and break
    while True:
        start_date_str = input("Enter the start date (YYYY-MM-DD HH:MM): ")
        try:
            start_date = datetime.strptime(start_date_str, '%Y-%m-%d %H:%M')
            break  # Exit the loop if the date is valid
        except ValueError:
            print("Invalid date format. Please try again.")

    # Daily cost input with validation using while loop, try and break
    while True:
        try:
            daily_cost = float(input("Enter the daily cost associated with this habit: "))
            if daily_cost < 0:
                print("Daily cost cannot be negative. Please enter a valid cost.")
                continue
            break  # Exit the loop if the input is valid
        except ValueError:
            print("Invalid input. Please enter a numeric value for daily cost.")

    # Minutes wasted input with validation
    while True:
        try:
            minutes_wasted = int(input("Enter the minutes wasted on this habit daily: "))
            if minutes_wasted < 0:
                print("Minutes wasted cannot be negative. Please enter a valid number.")
                continue
            break  # Exit the loop if the input is valid
        except ValueError:
            print("Invalid input. Please enter an integer value for minutes wasted.")

    # Add the habit to the list
    habits.append(breaking_habit(name_of_habit, start_date, daily_cost, minutes_wasted, wage_per_hour))

# Create DataFrame for created list
df = pd.DataFrame(habits)

# Print results with Tabulate, including username
print("\nPersonal Habit Tracker Summary for:", username)
print(tabulate(df, headers='keys', tablefmt='fancy_grid'))

```
